```python
import matplotlib.pyplot as plt
import json
import math
import pandas as pd
import numpy as np
```

```python
with open("brain.json") as fd:
    data = json.load(fd)
```

```python
N = len(data["result"])

tmp = []

running_score = {"win": 0, "loss": 0, "draw": 0}

for ii in range(N):
    result = data["result"][ii]
    running_score[result] += 1
    score = data["score"][ii]
    gen = data["generation"][ii]
    turns = data["turns"][ii]
    
    res = {
        "i": ii,
        "score": score,
        "gen": gen,
        "turns": turns,
        **running_score
    }
    tmp.append(res)

df = pd.DataFrame(tmp)
```

```python
def moving_average(a, n=5):
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
avg = moving_average(np.array(df["score"]))
plt.plot(range(len(avg)), avg)
```

```python
fig, ax = plt.subplots()
df.plot(x="i", y="win", ax=ax)
df.plot(x="i", y="draw", ax=ax)
df.plot(x="i", y="loss", ax=ax)
```
